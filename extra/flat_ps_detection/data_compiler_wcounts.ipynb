```python
# produce approximate histogram from point sources

nc = 40
index_counts = np.where(counts_list == nc)[0][0]

# load data and locations of point sources
username="ramirez"
local_dir = "/het/p4/"+username+"/gcewavelets/cwt_v0.1_play/"
results_dir = local_dir + 'results/processed/skymaps_sps_' + str(nc) + '/'

cwt_map_wps = np.load(results_dir + 'cwt_map.npy', allow_pickle = True)
xloc_list, yloc_list = np.load( 
    ("data/skymap/maps_by_counts/" + "skymaps_sps_" + str(nc) + '/' + 'ps_locs.npy')
     , allow_pickle=True)
loc_list = np.vstack((xloc_list, yloc_list)).T

# define mesh grid
mesh_bx_nb, mesh_by_nb = np.meshgrid(arr_bx_nb,arr_by_nb) # each output array (NxN shaped) contains x or y value at given (i,j)-th position
mesh_bxby_nb = np.stack((mesh_bx_nb, mesh_by_nb), axis=-1)

def closest_grid_point_(point, grid_points):
    relative_positions = grid_points - point
    distances = np.linalg.norm(relative_positions, axis = -1)
    bxby_index_shape = distances.shape
    flattened_index = np.argmin(distances)
    return np.unravel_index(flattened_index, bxby_index_shape)
```

```python
# produce approximate histogram from point sources

psf_scale = 0.4 / 15 / 2
effective_ps_size = 2 * psf_scale

N_maps = 1000

# for n_map in range(N_maps):
#     buf_loc_list = loc_list[np.newaxis, np.newaxis]
#     point = loc_list[n_map]
#     buf_point = point[np.newaxis, np.newaxis]
    
#     index = closest_grid_point_(point, mesh_bxby_nb_nb)
#     nx = index[0]
#     ny = index[1]

#     closest_point = mesh_bxby_nb[index]
#     buf_closest_point = closest_point[np.newaxis, np.newaxis]
#     relative_positions = buf_closest_point - mesh_bxby_nb
#     distances = np.linalg.norm(relative_positions, axis = -1)
#     indices = np.where(distances < effective_ps_size)
#     kx = indices[0]
#     ky = indices[1]
#     kxky = np.vstack((kx, ky)).T
    
#     if n == 0:
#         kx_list = kx
#         ky_list = ky
#         kxky_list = kxky
    
#     else:
#         kx_list = np.concatenate((kx_list, kx))
#         ky_list = np.concatenate((ky_list, ky))
#         kxky_list = np.vstack((kxky_list, kxky))
        
#     _, unique_indices = np.unique(kxky_list, axis = 0, return_index = True)
#     unique_kx = kx_list[unique_indices]
#     unique_ky = ky_list[unique_indices]

#     cwt_sig = cwt_map_wps[unique_ky,unique_kx,na]

cwt_map_wps_nb = np.zeros((Nx_nb, Nx_nb, N_maps))
for nx in range(Nx_nb):
    kx = n_nb[nx]
    for ny in range(Nx_nb):
        ky = n_nb[ny]
        cwt_map_wps_nb[nx,ny,:] = cwt_map_wps[kx,ky,0,:]
        
cwt_sig = np.empty(shape = (1))
psf_scale = 0.4 / 15 / 2
effective_ps_size = 0.05*psf_scale

for n_map in range(N_maps):
    buf_loc_list = loc_list[np.newaxis, np.newaxis]
    point = loc_list[n_map]
    buf_point = point[np.newaxis, np.newaxis]
    
    index = closest_grid_point_(point, mesh_bxby_nb)
    nx = index[0]
    ny = index[1]

    closest_point = mesh_bxby_nb[index]
    buf_closest_point = closest_point[np.newaxis, np.newaxis]
    relative_positions = buf_closest_point - mesh_bxby_nb
    distances = np.linalg.norm(relative_positions, axis = -1)
    indices = np.where(distances < effective_ps_size)
    kx = indices[0]
    ky = indices[1]

    cwt_sig_single = cwt_map_wps_nb[ky,kx,n_map].T
    cwt_sig = np.concatenate((cwt_sig, cwt_sig_single))
```
